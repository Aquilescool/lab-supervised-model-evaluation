```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from scipy import stats
from sklearn import linear_model
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## Regression Model Evaluation

```python
from sklearn.datasets import load_boston
data = load_boston()
X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.DataFrame(data["target"], columns=['MEDV'])

data = pd.concat([X, y], axis=1)
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python

model=LogisticRegression(random_state=0).fit(X_train, y_train.values.ravel().astype('int'))

y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
from sklearn.metrics import r2_score
print(r2_score(y_train, y_pred_train))
print(r2_score(y_test, y_pred_test))
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_train, y_pred_train))
print(mean_squared_error(y_test, y_pred_test))
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_train, y_pred_train))
print(mean_absolute_error(y_test, y_pred_test))
```

## Classification Model Evaluation

```python
from sklearn.datasets import load_iris

data = load_iris()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.DataFrame(data["target"], columns=["class"])

data = pd.concat([X, y], axis=1)
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
model=LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)
y_pred_train
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
from sklearn.metrics import classification_report
from sklearn import metrics

print('Accuracy : {:.2f}%'.format(metrics.accuracy_score(y_pred_train, y_train)*100))
print('Accuracy : {:.2f}%'.format(metrics.accuracy_score(y_pred_test, y_test)*100))
```
